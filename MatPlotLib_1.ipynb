```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
x = np.linspace(0, 10, 50)
plt.plot(x, np.sin(x))   # plot(x,y)
z=plt.plot(x, np.cos(x))
plt.show()
```

```python
fig = plt.figure()   # a container containing all tools to plot like text, graphics, axes etc(dosent create a figure)
ax = plt.axes()      # from the fig container we get axes which is ploted below
```

```python
plt.figure() # create a plot figure                       # MATLAB INTERFACE TO PLOT
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# ax will be an array of two Axes objects        # OBJECT ORIENTED INTERFACE TO PLOT
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-1,20,100)
plt.plot(x,np.sin(x), linestyle='-.', color='yellow')
plt.xlim(-1,14)
plt.ylim(-2,2)
```

```python
x=np.linspace(-1,20,100)
plt.plot(x,np.sin(x), linestyle='-.', color='yellow')
plt.xlim(-1,14)
plt.ylim(-2,2)
plt.title('Sine Wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')  #matches resolution to screens resolution
plt.axis('tight')  #stretches graph for good visibililty
plt.legend();      #makes legend
```
