```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.legend()
plt.axis('equal')
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.legend(loc='upper left', frameon=False) #turns frame for legend off and moves the legend
plt.axis('equal')                           #to top left as compared to above figure
```
